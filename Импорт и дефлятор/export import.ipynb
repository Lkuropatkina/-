```python
def det_import_export(num):
    path_name1 = "Eora\Eora26_20" + num + "_bp\\"
    
    labels_T = pd.read_table(path_name1 + "labels_T.txt", header=None)
    labels_T = labels_T.drop(4, axis=1) # был какой-то левый нулевой столбец
    labels_FD = pd.read_table(path_name1 + "labels_FD.txt", header=None)
    labels_FD = labels_FD.drop(4, axis=1) # был какой-то левый нулевой столбец
    
    path_name2 = path_name1 + "import\miranda\Eoras\Phase199\Loop082\_\simplified\\" 
    T = pd.read_table(path_name2 + "Eora26_20"+num+"_bp_T.txt", header=None)
    T.columns = labels_T[0]
    
    FD = pd.read_table(path_name2 + "Eora26_20"+num+"_bp_FD.txt", header=None)
    FD.columns = labels_FD[0]
    
    T = pd.concat([labels_T[0], T], axis=1, sort=False)
    FD = pd.concat([labels_T[0], FD], axis=1, sort=False)
    
    T = T.groupby([0]).sum()
    T = T.groupby(T.columns, axis = 1).sum()
    np.fill_diagonal(T.values, 0)
    
    FD = FD.groupby([0]).sum()
    FD = FD.groupby(FD.columns, axis = 1).sum()
    np.fill_diagonal(FD.values, 0)
    
    data = T+FD
    my_export = data.sum(axis = 1)
    my_import = data.sum()
    
    my_export = pd.DataFrame(my_export)
    my_import = pd.DataFrame(my_import)
    
    my_export["year"] = "20" + num
    my_import["year"] = "20" + num
    
    #my_import.to_excel('import_20'+ num +'.xlsx', index = False, encoding = 'utf-8')
    #my_export.to_excel('export_20'+ num +'.xlsx', index = False, encoding = 'utf-8')
    
    print(num + " год пройден")
    
    return my_export, my_import
```

```python
# i это год. Число от 0 до 15 
all_export = pd.DataFrame()
all_import = pd.DataFrame()

for i in range(16):
    num = format(i, '02d')
    res = det_import_export(num)
    all_export = pd.concat([all_export, res[0]])
    all_import = pd.concat([all_import, res[1]])
```

```python
all_export = all_export[["year", 0]]
all_export = all_export.sort_index()
all_export["country"] = all_export.index
all_export
```

```python
all_export = all_export.sort_values(by=['country','year'])
all_export
```

```python
all_import = all_import[["year", 0]]
all_import = all_import.sort_index()
all_import["country"] = all_import.index
all_import
```

```python
all_import = all_import.sort_values(by=['country','year'])
all_import
```

```python
all_import.to_excel('all_import.xlsx', index = False, encoding = 'utf-8')
all_export.to_excel('all_export.xlsx', index = False, encoding = 'utf-8')
```
